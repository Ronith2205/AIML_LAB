```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y1=[0,20,80,90,30];
y2=[10,40,70,60,80]
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("year")
plt.ylabel("grate")
plt.show()
```
